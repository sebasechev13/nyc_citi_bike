```python
import glob
import pandas as pd 
```

```python
# Merging csv files (without repeating header)
all_years_files = glob.glob("data/*.csv") 
df = pd.concat((pd.read_csv(f, header = 0) for f in all_years_files))
df
```

```python
# Export file as a CSV
df.to_csv("data/2019-citibike-tripsdata.csv")
```
